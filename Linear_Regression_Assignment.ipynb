## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
from sklearn.model_selection import train_test_split
```

```python
# Input Data
x = boston.data
  
# Output Data
y = boston.target
```

```python
#or

# Input Data
x=df.loc[:, df.columns != 'prices']
  
# Output Data
y=df.loc[:, df.columns == 'prices']
```

```python
#or
x = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = df['prices']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3, random_state=23)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model = LinearRegression()
```

## Q9: Train the Model.

```python
lin_model.fit(X_train, Y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(lin_model.intercept_) 
```

```python
coeff_df = pd.DataFrame(lin_model.coef_,x.columns,columns=['Coefficient'])
coeff_df # To see how the price will be affected when we make all fixed and increase one of Coefficient
```

```python
predictions = lin_model.predict(X_test)
predictions
```

```python
Real_Values = np.array(Y_test)
Real_Values
```

```python
plt.scatter(Real_Values,predictions); 
```

## Residual Histogram

```python
sns.distplot((Y_test-predictions),bins=50);
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(Y_test, predictions))
print('MSE:', metrics.mean_squared_error(Y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, predictions)))
```

```python
r2 = metrics.r2_score(Y_test, predictions)
r2
```
